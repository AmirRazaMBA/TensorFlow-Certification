# Keras Available Datasets

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
```

## MNIST

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data(path="mnist.npz")
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((x_train, y_train))
```

```python
train_dataset.cache().batch(32).shuffle(1000)
```

```python
test_dataset = tf.data.Dataset.from_tensor_slices((x_test, y_test))
```

```python
test_dataset.size()
```

```python
for image, label in train_dataset:
    #plt.imshow(image)
    print(label)
```

```python
for i in train_dataset:    
    print(i)
#print(i.numpy())
```

```python
n = 10
print(plt.imshow(x_train[n]))
print(y_train[n])
```

changing tuples to dataset and then iterating  

```python
train_ds = tf.data.Dataset.from_tensor_slices((x_train,y_train))
```

```python
type(train_ds)
```

```python
# iteration method : 1
for image, label in train_ds.take(1):
    plt.imshow(image)
    print(label)
```

```python
# iteration method : 2
it = next(iter(train_ds))
image, label = it
plt.imshow(image)
print(label.numpy)
```
